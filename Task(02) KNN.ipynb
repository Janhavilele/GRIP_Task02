# Finding optimum number of cluster for k-means classification

```python
x= data.iloc[:,[1,2,3]].values

from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters= i , init= 'k-means++',max_iter=300 , n_init = 10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

# plotting results onto one graph

```python
plt.plot(range(1,11),wcss,marker="o")
plt.title('The elbow method')
plt.ylabel('wcss')
plt.xlabel('Number of clusters')
plt.show()
```

from above graph we can clearly say that our elbow point is at k=3

# Applying KMeans algorithm

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
y_kmeans
```

```python
# Checking the cluster of each species
data["cluster"] = y_kmeans
data
```

```python
# finding the centroid for each cluster
centroid = kmeans.cluster_centers_
print(centroid)
```

# Visualizing the clusters

```python
# Visualising the clusters - On the first two columns
plt.figure(figsize=(8,6))
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
